# Rectified Linear Unit- Relu 

f(x)=max(0,x)


```python
import numpy as np
from numpy import *
import matplotlib.pylab as plt
%matplotlib inline

'''
作業:
    寫出 ReLU & dReLU 一階導數
    並列印
'''
```

```python
def ReLU(x):
    return abs(x) * (x > 0)

def dReLU(x):
    return 1 * (x > 0)
```

```python
x = np.linspace(-10, 10, 100)

plt.plot(x, ReLU(x),'r')
plt.title('ReLU Function')
plt.text(-8, 8, 'f(x)=max(0,x)', fontsize=15,)
plt.grid()
plt.show()
```

```python

x = np.linspace(-10, 10, 100)
plt.plot(x, dReLU(x),'b')
plt.title('dReLU ')
plt.text(0.6,0.7 , 'f(x)\'=max(0,x)', fontsize=15)
plt.grid()
plt.show()
```

```python
x = np.linspace(-3, 3, 100)

plt.title('ReLU & dReLU Function')
plt.plot(x, ReLU(x),'r')
plt.plot(x, dReLU(x),'b')
#plt.title('dReLU ')
plt.text(-1,2 , 'ReLU:f(x)=max(0,x)', {'color': 'r', 'fontsize': 15})
plt.text(-3,1.2, 'dReLU:f(x)\'=max(0,x)',  {'color': 'b', 'fontsize': 15})
plt.grid()
plt.show()
```

```python
def ReLU1(x):
    return list(map(lambda x:x if x>0 else 0,x))
  
def dReLU1(x):
    return list(map(lambda x: 1 if x>0 else 0,x))
```

```python
x = np.linspace(-10, 10, 100)

plt.plot(x, ReLU1(x),'r')
plt.title('ReLU Function')
plt.text(-8, 8, 'f(x)=max(0,x)', fontsize=15,)
plt.grid()
plt.show()
```

```python

x = np.linspace(-10, 10, 100)
plt.plot(x, dReLU1(x),'b')
plt.title('dReLU ')
plt.text(0.6,0.7 , 'f(x)\'=max(0,x)', fontsize=15)
plt.grid()
plt.show()
```

```python
x = np.linspace(-3, 3, 100)

plt.title('ReLU & dReLU Function')
plt.plot(x, ReLU1(x),'r')
plt.plot(x, dReLU1(x),'b')
#plt.title('dReLU ')
plt.text(-1,2 , 'ReLU:f(x)=max(0,x)', {'color': 'r', 'fontsize': 15})
plt.text(-3,1.2, 'dReLU:f(x)\'=max(0,x)',  {'color': 'b', 'fontsize': 15})
plt.grid()
plt.show()
```
